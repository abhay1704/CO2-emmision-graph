```python
import pandas as pd
import numpy as np 
import json
```

```python
df = pd.read_excel('Data\Countrywise_yearwise_co2emission.xlsx')
df.head()
```

```python
df_filtered = df.dropna(axis=0)
df_filtered.shape, df.shape
```

```python
df_filtered.drop(['Indicator Name',	'Indicator Code'], axis=1, inplace=True)
df_filtered.head()
```

```python
data = []

# df_filtered.set_index('Country Name', inplace=True)

for country in df_filtered.index:
    countryData = {
        "countryName": country,
        "countryCode" : df_filtered.loc[country]['Country Code']
    }
    for year in range(1990, 2021):
        countryData[year] = df_filtered.loc[country][year]

    data.append(countryData)

co2_per_capita = {
    'About' : 'CO2 emissions (metric tons per capita)',
    'dataSource' : {
        'name' : 'Data world bank',
        'link' : 'https://data.worldbank.org/indicator/EN.ATM.CO2E.PC'
    },
    'data':data,
}
```

```python
import json
```

```python
json.dump(co2_per_capita, open('json/countrywise_co2PC.json', 'w'))
# help(json)
```
